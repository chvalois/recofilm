```python
df_content['genres'] = df_content['genres'].str.split(', ')
df_content['tag'] = df_content['tag'].str.split(', ')
df_content.head()
```

```python
def unique_tags(x):
    unique_values = set()
    for sublist in x:
        unique_values.update(sublist)
    return list(unique_values)
```

```python
df_unique_tags = df_content.groupby('movieId')['tag'].apply(unique_tags).reset_index()
df_unique_tags.head()
```

```python
df_content = df_content.drop(["tag"], axis=1)
df_content.head()
```

```python
df_content_tags = df_content.drop_duplicates(subset=['movieId']).merge(df_unique_tags, on='movieId', how='left')
df_content_tags
```

```python
unique_movieIds_count = df_content_tags['movieId'].nunique()
print("Nombre de movieId uniques :", unique_movieIds_count)
```

```python
count_nan_tags = df_content_tags['tag'].apply(lambda x: 'nan' in x).sum()
print("Nombre de lignes avec 'nan' dans la liste de tags :", count_nan_tags)
```

```python
df_content_tags['combined'] = df_content_tags.apply(lambda row: row['genres'] + row['tag'] if row['tag'] != ['nan'] else row['genres'], axis=1)

df_content_tags = df_content_tags.drop(["genres","tag"], axis=1)
df_content_tags.head()
```

```python
df_content_tags.rename(columns={'combined': 'tags'}, inplace=True)
```

```python
df_content_tags.head()
```

```python
def remove_duplicates(lst):
    lst_lower = [tag.lower() for tag in lst]
    unique_tags = []
    seen = set()
    for tag in lst_lower:
        if tag not in seen:
            unique_tags.append(tag)
            seen.add(tag)
    
    return unique_tags

df_content_tags['tags'] = df_content_tags['tags'].apply(remove_duplicates)
```
